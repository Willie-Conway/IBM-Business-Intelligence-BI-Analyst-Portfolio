## **Transaction Trends Analysis**
# 
### This notebook visualizes transaction trends over time using sample data.

```python

# Import required libraries
import matplotlib.pyplot as plt
import numpy as np

# Set matplotlib style for better visual appearance
plt.style.use('default')

# %%
# Define the data
months = ['January 2023', 'February 2023']
transactions = [30, 30]

# %%
# Create the visualization
plt.figure(figsize=(10, 5))
plt.plot(months, transactions, marker='o', linewidth=2, markersize=8, color='steelblue')

# Customize the plot
plt.title('Transaction Trends Over the Year', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)

# Add value labels on each point
for i, (month, value) in enumerate(zip(months, transactions)):
    plt.text(i, value + 0.5, str(value), ha='center', fontsize=10)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()

# %%
# Create a bar chart alternative for comparison
plt.figure(figsize=(10, 5))
bars = plt.bar(months, transactions, color=['skyblue', 'lightcoral'], edgecolor='black')

# Customize the bar chart
plt.title('Transaction Count by Month', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.3,
             f'{height}', ha='center', fontsize=10)

plt.tight_layout()
plt.show()

# %%
# Data summary
print("=== Transaction Data Summary ===")
print(f"Months analyzed: {len(months)}")
print(f"Total transactions: {sum(transactions)}")
print(f"Average transactions per month: {np.mean(transactions):.2f}")
print(f"Minimum transactions: {min(transactions)}")
print(f"Maximum transactions: {max(transactions)}")
```
