```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

skills = """
Python Python Python Python
SQL SQL SQL SQL
JavaScript JavaScript
TypeScript TypeScript
Java Java
C++ C++
PHP PHP
Kotlin Kotlin
Swift Swift
R R

Pandas Pandas
NumPy NumPy
Scikit-Learn Scikit-Learn
SciPy SciPy
Statsmodels Statsmodels

Matplotlib Matplotlib
Seaborn Seaborn
Plotly Plotly
Folium Folium
Tableau Tableau

Dash Dash
Jupyter Jupyter
Streamlit Streamlit
React React

SQLite SQLite
MySQL MySQL
PostgreSQL PostgreSQL
MongoDB MongoDB
Cassandra Cassandra
MariaDB MariaDB
DynamoDB DynamoDB

Apache_Spark Apache_Spark
Apache_Hadoop Apache_Hadoop
Apache_Airflow Apache_Airflow
Apache_Kafka Apache_Kafka

Power_BI Power_BI
IBM_Cognos IBM_Cognos
Google_Looker Google_Looker
Google_Data_Studio Google_Data_Studio

HTML5 HTML5
CSS3 CSS3
Django Django
Angular Angular
Vue.js Vue.js
Node.js Node.js

Network_Security Network_Security
Risk_Management Risk_Management
Google_Cybersecurity Google_Cybersecurity

Docker Docker Docker
AWS AWS AWS
Kubernetes Kubernetes Kubernetes
Google_Cloud Google_Cloud
Azure Azure
GitHub_Actions GitHub_Actions
Linux Linux Linux

Git Git
GitHub GitHub
Jenkins Jenkins
Apache Apache
Nginx Nginx

Figma Figma
Adobe_Photoshop Adobe_Photoshop
Canva Canva
Google_UX_Design Google_UX_Design

Google_BI Google_BI
Google_Data_Analytics Google_Data_Analytics
Google_Advanced_Analytics Google_Advanced_Analytics

Google_IT_Support Google_IT_Support
Python_Automation Python_Automation
Google_Project_Management Google_Project_Management
Google_Digital_Marketing Google_Digital_Marketing

Meta_Software_Engineer Meta_Software_Engineer
Meta_Full_Stack Meta_Full_Stack
Meta_Data_Analyst Meta_Data_Analyst
Meta_Front_End Meta_Front_End
Meta_Back_End Meta_Back_End
Meta_Database_Engineer Meta_Database_Engineer
"""

wordcloud = WordCloud(
    width=1800,
    height=900,
    background_color="white",
    colormap="viridis",
    prefer_horizontal=0.9,
    collocations=False
).generate(skills)

plt.figure(figsize=(16,8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
output_file = "full_skills_wordcloud.png"
wordcloud.to_file(output_file)

print(f"Saved as {output_file}")
```
